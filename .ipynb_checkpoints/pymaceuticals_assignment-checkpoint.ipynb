```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_mouse_df = pd.merge(study_results, mouse_metadata, how="left", on=['Mouse ID', 'Mouse ID'])

# Display the data table for preview
study_mouse_df
```

```python
# Checking the number of mice.
mouse_count = len(study_mouse_df['Mouse ID'].unique())

mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates_mice = study_mouse_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)

# Locate all the duplicated mice and return Mouse ID
duplicates_mice_id = study_mouse_df.loc[duplicates_mice, 'Mouse ID'].unique()

duplicates_mice_id 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicates_mice_all = study_mouse_df.loc[study_mouse_df['Mouse ID']==duplicates_mice_id[0]]
duplicates_mice_all
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_count_df = study_mouse_df.loc[study_mouse_df['Mouse ID']!='g989']

clean_mouse_count_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_count2 = len(clean_mouse_count_df['Mouse ID'].unique())
clean_mouse_count2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
drug_regimen = study_mouse_df.groupby('Drug Regimen')

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_vol_mean = drug_regimen['Tumor Volume (mm3)'].mean()
tumor_vol_median = drug_regimen['Tumor Volume (mm3)'].median()                   
tumor_vol_variance = drug_regimen['Tumor Volume (mm3)'].var()                     
tumor_vol_stdev = drug_regimen['Tumor Volume (mm3)'].std()                   
tumor_vol_sem = drug_regimen['Tumor Volume (mm3)'].sem()
                 
# Assemble the resulting series into a single summary DataFrame.
summary_stats_df = pd.DataFrame({'Mean Tumor Volume':tumor_vol_mean,
                                 'Median Tumor Volume':tumor_vol_median,
                                'Tumor Volume Variance':tumor_vol_variance,
                                'Tumor Volume Std. Dev.':tumor_vol_stdev,
                                'Tumor Volume Std. Err.':tumor_vol_sem
                                })
                 
# summary_stats_df = summary_stats.set_index('Drug Regime') check if this is required qith groupby
summary_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
summary_stats2 = drug_regimen['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

summary_stats2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
final_tumor = clean_mouse_count_df.loc[(clean_mouse_count_df['Drug Regimen']=='Capomulin')|
                                       (clean_mouse_count_df['Drug Regimen']=='Ramicane')|
                                       (clean_mouse_count_df['Drug Regimen']=='Infubinol')|
                                       (clean_mouse_count_df['Drug Regimen']=='Ceftamin')]

# Start by getting the last (greatest) timepoint for each mouse
final_tumor_max_tp = final_tumor.groupby(['Drug Regimen','Mouse ID'])['Timepoint'].max().reset_index()

final_tumor_max_tp
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_vol_tp = pd.merge(final_tumor_max_tp, clean_mouse_count_df[['Mouse ID','Timepoint','Tumor Volume (mm3)']],
                                                                        on=['Mouse ID','Timepoint'], how='left')

# Keep the df seperate for later
final_tumor_vol_tp_df = final_tumor_vol_tp

final_tumor_vol_tp_df
```

```python
print(final_tumor_vol_tp_df.columns.tolist())
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_regimen = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []
    
# Calculate the IQR and quantitatively determine if there are any potential outliers.   
for treatment in drug_regimen: 
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol_tp = final_tumor_vol_tp_df.loc[final_tumor_vol_tp_df['Drug Regimen'] == treatment,
                                                       'Tumor Volume (mm3)']
     
    # add subset
    tumor_vol_data.append(final_tumor_vol_tp)    
    
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = final_tumor_vol_tp.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    mean = final_tumor_vol_tp.mean()
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    # Determine outliers using upper and lower bounds
    outliers = final_tumor_vol_tp.loc[(final_tumor_vol_tp < lower_bound)|(final_tumor_vol_tp > upper_bound)]
    
    
    #Print results for comparison and visual information for plots
    print(f' ')
    print(f'{treatment} Regimen')
    print(f'The Outlier of Tumor Volume(mm3) for {treatment} could be {round(lower_bound,2)}')
    print(f'The Outlier of Tumor Volume(mm3) for {treatment} could be {round(upper_bound,2)}')
    print(f'The Number of outliers for Capomulin is: {outliers.count()}')
    print(f"The Lower Quartile of Tumor Volume is: {lowerq}")
    print(f"The Upper Quartile of Tumor Volume is: {upperq}")
    print(f"The Interquartile Range of the Tumor Volume is: {iqr}")
    print(f"The Median of the Tumor Volume is: {quartiles[0.5]} ")
    print(f'The Average Tumor Volume(mm3) for {treatment} is: {round(mean,2)}')
    if outliers.count() > 0:
        for i in outliers.index:
            print(f'Outlier for the Final Tumor Volume(mm3) is {round(outliers[i], 2)}')
```

```python
# Check the tumor data is collated correctly
tumor_vol_data
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig, tumor_bxplot = plt.subplots()
tumor_bxplot.boxplot(tumor_vol_data, labels = drug_regimen, flierprops = {'markerfacecolor':'r','markersize': 12})

tumor_bxplot.set_xlabel('Drug Regimen')
tumor_bxplot.set_ylabel('Final Tumor Volume (mm3)')
# tumor_bxplot.set_title('Distribution of Tumor Volume for Treatment Groups')

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
#  Capomjulin DataFrame
capomulin = clean_mouse_count_df[clean_mouse_count_df['Drug Regimen']=='Capomulin']

# Obtain the averages for plotting
tumor = capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weight = capomulin.groupby('Mouse ID')['Weight (g)'].mean()

# Create Scatter Plot
plt.scatter(weight, tumor)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
# plt.title('Mouse Weight vs. Average Observed Tumor Volume for Capomulin Treatment')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# Import scipy.stats to complete linregress
from scipy.stats import linregress

x_values = (weight)
y_values = (tumor)

# Print the correlation above the Graph
correlation = round(st.pearsonr(x_values, y_values)[0],2)
print(f'The correlation between mouse weight and tumor vol is {correlation}')

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = 'y = '' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))

plt.scatter(x_values,y_values)

plt.plot(x_values,regress_values,'r-')
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color='red')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
# plt.title('Mouse Weight vs. Average Observed Tumor Volume for Capomulin Treatment Correalation & Linear Regression')

plt.show()
```
